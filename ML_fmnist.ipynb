```python
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
fmnist = tf.keras.datasets.fashion_mnist
import numpy as np
from tensorflow.keras.layers import Dropout
```

```python
(x_train, y_train), (x_valid, y_valid) = fmnist.load_data()
```

```python
print(x_train.shape)
print(x_valid.shape)
print(x_train.min())
print(x_train.max())

x_train = x_train.reshape(60000,784)
x_valid = x_valid.reshape(10000,784)
x_train = x_train/255
x_valid = x_valid/255

x_train = np.array(x_train)
x_valid = np.array(x_valid)

y_train = np.array(y_train)
y_valid = np.array(y_valid)
```

```python
# print(y_train)
# print(y_valid[0:20])
# print(y_train.shape)
# print(y_valid.shape)
num_categories = 10

y_train = keras.utils.to_categorical(y_train, num_categories)
y_valid = keras.utils.to_categorical(y_valid, num_categories)
```

```python
my_first_AI = Sequential()
```

```python
my_first_AI.add(Dense(units=400, activation='relu')) # input layer
my_first_AI.add(Dense(units=400, activation='relu')) #first hidden layer
my_first_AI.add(Dropout(0.1))
my_first_AI.add(Dense(units = 10, activation='softmax')) #output layer
```

```python
my_first_AI.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
ep_ic = 5
epic_games = my_first_AI.fit(x_train, y_train, epochs=ep_ic, validation_data=(x_valid, y_valid))
```

```python
my_first_AI.summary()
print(my_first_AI.optimizer.learning_rate.numpy())
```
